# Reviews Analyzer App

```python
# let's build the Reviews Analyzer using Transformer pipeline 
from transformers import pipeline
import pandas as pd

# Load pre-trained sentiment-analysis pipeline
sentiment_pipeline = pipeline(
                    task="sentiment-analysis", 
                    model="distilbert-base-uncased-finetuned-sst-2-english", 
                    framework="pt"
)
input_reviews = [
    "I absolutely love this product! It has changed my life for the better.",
    "This is the worst purchase I've ever made. Completely disappointed.",
    "It's okay, not great but not terrible either.",
    "I can't believe how much I love this product. Highly recommended!"    
]
# Analyze sentiments
results = sentiment_pipeline(input_reviews)
results_df = pd.DataFrame(results)
print(results_df)
```
